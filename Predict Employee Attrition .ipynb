# Step 2: Data Preprocessing & Model Training Preparation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, roc_auc_score, roc_curve
import shap
```

*2.1 Target variable encoding*

```python
le = LabelEncoder()
df['Attrition'] = le.fit_transform(df['Attrition']) # Yes=1, No=0
print("\nAttrition column encoded (0=No, 1=Yes):")
print(df['Attrition'].value_counts())
```

*2.2 Define Features (X) and Target (y)*


```python
X = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
# --- 2.3 Identify Numerical and Categorical Features (Again, after drops) ---
numerical_cols = X.select_dtypes(include=np.number).columns.tolist()
categorical_cols = X.select_dtypes(include='object').columns.tolist()
```

```python
# --- 2.4 Create Preprocessing Pipelines ---
# Pipeline for numerical features: Scale them
numerical_transformer = StandardScaler()
```

```python
# Pipeline for categorical features: One-Hot Encode them
categorical_transformer = OneHotEncoder(handle_unknown='ignore', drop='first') # drop='first' to avoid multicollinearity
```

```python
# Combine preprocessing steps using ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ],
    remainder='passthrough' # Keep any columns not specified (shouldn't be any here)
)
```

```python
# --- 2.5 Train-Test Split ---
# Stratify ensures the proportion of attrition is similar in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print("\nTraining set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

# Step 3: Training Classification Models

```python
# --- 3.1 Logistic Regression ---
# Create the full pipeline: Preprocessing + Model
lr_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', LogisticRegression(solver='liblinear', class_weight='balanced', random_state=42))]) # Added class_weight
```

```python
# Train the model
print("\nTraining Logistic Regression...")
lr_pipeline.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred_lr = lr_pipeline.predict(X_test)
y_prob_lr = lr_pipeline.predict_proba(X_test)[:, 1] # Probabilities for ROC AUC
```

```python
# Evaluate
print("\nLogistic Regression Evaluation:")
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("ROC AUC:", roc_auc_score(y_test, y_prob_lr))
print("Classification Report:\n", classification_report(y_test, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
```

# Step 4: Use SHAP to Explain Model Predictions

**Explaining Logistic Regression with SHAP**

```python
# Get the fitted preprocessor and logistic regression model from the pipeline
preprocessor = lr_pipeline.named_steps['preprocessor']
model_lr = lr_pipeline.named_steps['classifier']
```

```python
# applying the fitted preprocessor to the training and test sets
X_train_processed = preprocessor.transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
# Get feature names after one-hot encoding
try:
    # Attempt to get feature names from the 'cat' transformer
    ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)
except AttributeError:
     # Fallback for older scikit-learn versions might be needed,
     # but get_feature_names_out is standard now.
     # If this fails, manual creation based on categories might be required.
     print("Warning: Could not automatically retrieve OHE feature names. Ensure sklearn version is recent.")
     # As a basic fallback (less ideal):
     # ohe_feature_names = [f"cat_{i}" for i in range(X_train_processed.shape[1] - len(numerical_cols))]
     # Or reconstruct manually if needed based on the encoder's categories_ attribute.
     # For now, assume get_feature_names_out works.
     ohe_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(input_features=categorical_cols)
```

```python
# Combine numerical and OHE feature names in the correct order
feature_names_processed = numerical_cols + list(ohe_feature_names)
```

```python
# Converting the processed data back to DataFrames for easier use with SHAP plots
X_train_processed_df = pd.DataFrame(X_train_processed, columns=feature_names_processed)
X_test_processed_df = pd.DataFrame(X_test_processed, columns=feature_names_processed)
```

```python
print(f"Processed training data shape: {X_train_processed_df.shape}")
print(f"Processed test data shape: {X_test_processed_df.shape}")
print(f"Number of features after processing: {len(feature_names_processed)}")
```

**Initialize SHAP LinearExplainer and Calculate Values**

```python
# Use LinearExplainer for Logistic Regression.
# It needs the model and typically some background data (often the training data)
# to determine the expected value (baseline prediction).
print("\nInitializing SHAP LinearExplainer...")
explainer = shap.LinearExplainer(model_lr, X_train_processed_df, feature_perturbation="interventional")
```

```python
# Calculating SHAP values for the processed test set
print("Calculating SHAP values for the test set...")
shap_values = explainer.shap_values(X_test_processed_df)

print(f"SHAP values calculated. Shape: {shap_values.shape}") # Should be (n_test_samples, n_features)
```

**Visualize SHAP Values**

```python
# global Feature Importance (Bar Plot)
print("\nGenerating SHAP Summary Plot (Bar)...")
shap.summary_plot(shap_values, X_test_processed_df, plot_type="bar", show=False)
plt.title("SHAP Global Feature Importance (Logistic Regression)")
plt.show()
```

```python
# global Feature Importance and Impact (Beeswarm Plot - or 'dot' for LinearExplainer)
print("\nGenerating SHAP Summary Plot (Beeswarm/Dot)...")
shap.summary_plot(shap_values, X_test_processed_df, show=False)
```

```python
plt.title("SHAP Feature Impact on Attrition Prediction (Logistic Regression)")
plt.show()
```

```python
# Dependence Plots for Top Features
# getting features sorted by mean absolute SHAP value
abs_shap_mean = np.abs(shap_values).mean(axis=0)
feature_indices_sorted = np.argsort(abs_shap_mean)[::-1]
top_features = [feature_names_processed[i] for i in feature_indices_sorted]

print("\nGenerating SHAP Dependence Plots for Top Features (Logistic Regression)...")
for feature in top_features[:5]: # plot for top 5 features
    try:
        shap.dependence_plot(feature, shap_values, X_test_processed_df, interaction_index=None, show=False) # interaction_index=None is safer for Linear
        plt.title(f"SHAP Dependence Plot for {feature} (LR)")
        plt.show()
    except Exception as e:
        print(f"Could not generate dependence plot for {feature}: {e}")
        # Dependence plots might be less informative for OHE features; focus on numerical or interpretable categoricals if issues arise.
```

```python
# Explanation for a single prediction (Example: first prediction in test set)
print("\nGenerating SHAP Force Plot for a single prediction (Logistic Regression)...")
# Reload JS visualization library if in a suitable environment (like Jupyter)
try:
    shap.initjs()
except NameError: # Handle cases where shap.initjs() isn't needed/available
    pass
```

```python
instance_index = 0
# For LinearExplainer, expected_value is a single scalar
print(f"Base Value (Expected Prediction): {explainer.expected_value}")
shap.force_plot(explainer.expected_value, shap_values[instance_index,:], X_test_processed_df.iloc[instance_index,:], show=False)
plt.show() # May show static version or require interactive environment
```

*SHAP Interpretation:*

Summary Plot (Bar): Identifies the features with the largest average impact on the model's predictions (magnitude).

Summary Plot (Beeswarm): Shows not just the importance but how feature values affect the prediction. For example, you'll likely see:

OverTime_Yes dots concentrated on the right (high positive SHAP value), indicating working overtime strongly pushes the prediction towards attrition.

MonthlyIncome might show blue dots (low income) on the right and red dots (high income) on the left, indicating low income increases attrition risk.

TotalWorkingYears likely shows blue dots (low tenure) on the right and red dots (high tenure) on the left.

Dependence Plots: Reveal the relationship between a feature's value and its impact more clearly, potentially showing non-linearities or interaction effects.

# Step 5: Actionable Insights to Reduce Attrition

1.Address Overtime Culture:

- Insight: Working overtime (OverTime_Yes) is consistently a top predictor of attrition.

- Action: Investigate reasons for high overtime (workload, inefficiency, understaffing). Implement policies to limit mandatory overtime, promote work-life balance, assess team staffing levels, and potentially offer flexible schedules or compensation for extra hours.

2.Focus on Compensation and Growth for Junior/Lower-Income Roles:

- Insight: Low MonthlyIncome, low JobLevel, and sometimes low TotalWorkingYears strongly correlate with leaving.

- Action: Review salary benchmarks, especially for entry-level and lower job bands. Ensure compensation is competitive. Develop and clearly communicate career progression paths and opportunities for skill development and promotion. Implement performance-based bonuses or raises.

3.Improve Manager Relationships and Early Tenure Support:

- Insight: Low YearsWithCurrManager and YearsAtCompany are often linked to attrition. Low JobInvolvement or RelationshipSatisfaction might also appear.

- Action: Invest in manager training focused on people management, communication, and providing support. Implement structured onboarding programs. Encourage regular 1-on-1 meetings between managers and employees, especially newer ones. Foster team cohesion and positive relationships.

4.Enhance Job Satisfaction and Work Environment:

- Insight: Low JobSatisfaction and EnvironmentSatisfaction scores predict attrition. Specific JobRoles (e.g., Sales Rep, Lab Tech) might have higher rates.

- Action: Conduct targeted employee surveys or focus groups within high-attrition roles/departments to understand specific pain points. Address concerns related to the work environment, resources, recognition, and the nature of the work itself.

5.Targeted Retention for Specific Demographics:

- Insight: MaritalStatus_Single often shows higher attrition. Younger Age groups might be more prone to leaving. DistanceFromHome can play a role.

- Action: While harder to act on directly, understand the needs of these groups. For single employees, perhaps offer benefits or social activities that appeal to them. For younger employees, focus on growth opportunities and mentorship. For those with long commutes, explore remote/hybrid options if feasible.

6.Review Business Travel Policies:

- Insight: BusinessTravel_Travel_Frequently is often associated with higher attrition.

- Action: Evaluate the necessity of frequent travel. Offer alternatives like video conferencing where possible. Ensure adequate compensation, support, and time off for employees who travel frequently.
